## Part 0 — Computing environment mental model

Three places:
1. Host machine
2. Linux virtual machine
3. GitHub (remote)

Execution locations:
- Local (VM)
- Cloud (Codespaces)

Mental model:
> Code runs wherever the interpreter runs.


## Part 1 — What is programming?

Programming is specifying a finite sequence of precise instructions executable by a machine.

Key ideas:
- Precision
- Formality
- Executability


## Programming languages

A programming language is a formal system with:
- Syntax
- Semantics


## Syntax vs semantics

Syntax determines well‑formed programs.
Semantics determines behaviour.


## Values, expressions, statements

Value: data.
Expression: evaluates to a value.
Statement: performs an action or changes state.


## How programs run

Execution model:
1. Fetch instruction
2. Execute
3. Update state


## Compilation vs interpretation

Compiled:
- Translation before execution.

Interpreted:
- Execution via interpreter.

Hybrid:
- Compilation to intermediate form then execution.


## Python execution model

Python source → bytecode → virtual machine execution.


## Errors

Syntax error: invalid program.

Runtime error: failure during execution.


## State

Program behaviour depends on previous assignments.


## Environment model

Variables map to values.
Assignment updates mapping.
